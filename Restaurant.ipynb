```python
# -------------------------
# Class The class in Restaurant Management system ----
# ---------------------

# we are difining meanu like
# we use self variable to access the class local in current object .
# we declare MeanuItem and Order class global . when we need we can access .
# Declear MenuItem

class MenuItem:
    def __init__(self, item_id, name, price):
        self.item_id = item_id
        self.name = name
        self.price = price

# customer see the meanu and order the item the create order variable

class Order:
    def __init__(self, customer_name):
        self.customer_name = customer_name
        self.items = []
# cutomer add itme which item selected by user
    def add_item(self, item):
        self.items.append(item)
# total price of item order by cusotmer we use lamda because only one specfic context.
    def total(self):
        return sum(map(lambda x: x.price, self.items))
# order from customer show the list of items and price to the customer .
    def show_order(self):
        print(f"nOrder for {self.customer_name}:")
        for i, item in enumerate(self.items, 1):
            print(f"{i}. {item.name} - {item.price}")
        print(f"Total Amount: ₹{self.total()}")

#__________________________________________________________________
  # Resaurant management
  # make a class Restaurant a variable __init__(self) define it
  # call the data in menu variable

class Restaurant:
    def __init__(self):
        self.menu = []
        self.orders = []

    def load_menu(self):
        self.menu = [
            MenuItem(1, "Pizza", 200),
            MenuItem(2, "Burger", 100),
            MenuItem(3, "Fries", 70),
            MenuItem(4, "Coke", 40),
            MenuItem(5, "Pasta", 150)

        ]
#Peint the menu of restrant menu order by customer

    def show_menu(self):
        print(" MENU:")
        for item in self.menu:
            print(f"{item.item_id}. {item.name} - ₹{item.price}")
# customer select the item from menu where item_id = item
    def get_item_by_id(self, item_id):
        for item in self.menu:
            if item.item_id == item_id:
                return item
        return None

# show menu and take order from customer

    def take_order(self):
        customer_name = input("Enter customer name: ").title()
        order = Order(customer_name)
        self.show_menu()

#use while loop if cutomer put the order and item not exit in list then loop break it if item exxit add in card and and price if not exit invalid id

        while True:
            try:
                item_id = int(input("Enter item number (0 to finish): "))
                if item_id == 0:
                    break
                item = self.get_item_by_id(item_id)
                if item:
                    order.add_item(item)
                    print(f"Added {item.name}")
                else:
                    print("Invalid item ID.")
            except ValueError:
                print("Enter a valid number.")

# customer not selected any item prnt not select item if select show the item order by customer .

        if order.items:
            order.show_order()
            self.orders.append(order)
        else:
            print("No items selected.")
# show all order of cutomers
    def show_all_orders(self):
        if not self.orders:
            print("\nNo orders yet.")
            return
        print(" All Orders:")
        for order in self.orders:
            order.show_order()

# now load the input for cutomer and show the meanu to the customer for order and read the meanu cutomer
# and give the input to program choce the meanu if exit orer by cutomer then accept it or break the loop .

def main():
    restaurant = Restaurant()
    restaurant.load_menu()

    while True:
        print("-- Restaurant Management ---")
        print("1. Show Menu")
        print("2. Take New Order")
        print("3. View All Orders")
        print("4. Exit")

        choice = input("Choose an option: ")

        if choice == '1':
            restaurant.show_menu()
        elif choice == '2':
            restaurant.take_order()
        elif choice == '3':
            restaurant.show_all_orders()
        elif choice == '4':
            print("Goodbye! Visit again.")
            break
        else:
            print("❌ Invalid option. Try again.")

# Run the program
if __name__ == "__main__":
    main()
```
